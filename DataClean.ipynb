```python
df['name'] = df['Used name'].str.replace("•", " ")
df.head()
```

# Slit the Measurment column into height and weight column

```python
df[['Height/cm', 'Weight/kg']] = df['Measurements'].str.split('/', expand=True)
df.head()
```

```python
#Getting rid of "cm" and "kg" in the new columns
df['Height/cm'] = pd.to_numeric(df['Height/cm'].astype(str).str.strip(' cm'), errors = 'coerce')
df['Weight/kg'] = pd.to_numeric(df['Weight/kg'].astype(str).str.strip(' kg'), errors = 'coerce')
df.head()
```

```python
df.sort_values('Height/cm', ascending= False)
```

# Parse out 'Born' and 'Died' Column

```python
df['Born_date'] = df['Born'].str.extract(r'(\d+ \w+ \d{4}|\d{4})')
df['Born_date'] = pd.to_datetime(df['Born_date'], format="mixed", errors = 'coerce')

df['Died_date'] = df['Died'].str.extract(r'(\d+ \w+ \d{4}|\d{4})')
df['Died'] = pd.to_datetime(df['Died_date'], format="mixed", errors = 'coerce')
df['Died'] = df['Died'].astype(str)

df.head()
```

```python
df['Died'] = df['Died'].fillna('Alive').astype(str)
df.head()
```

# Get city, region and country from Born column

```python
pattern = r'in ([\w\s]+), ([\w\s]+) \((\w+)\)'
df[['Born_city', 'Born_region', 'Born_country']] = df['Born'].str.extract(pattern, expand= True)
df.head()
```

# Getting rid of extra columns

```python
df.columns
```

```python
important_columns = [
    'athlete_id', 'name', 'Born_date', 'Died_date',
    'Sex', 'NOC', 'Height/cm', 'Weight/kg',
    'Born_city', 'Born_region', 'Born_country'
]

df_clean = df[important_columns]
df_clean.head()
```

```python
df_clean.to_csv('new_bios.csv', index= False)
```
